```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc file CSV
file_path = "/content/drive/MyDrive/Python /subject-info.csv"
df = pd.read_csv(file_path)

# Hiển thị thông tin tổng quan
print("Thông tin tổng quan về dữ liệu:")
print(df.info())

# Hiển thị 5 dòng đầu tiên
print("\n5 dòng dữ liệu đầu tiên:")
print(df.head())

# Kiểm tra dữ liệu thiếu
print("\nSố lượng giá trị thiếu trong mỗi cột:")
print(df.isnull().sum())

# Vẽ biểu đồ phân bố giới tính
plt.figure(figsize=(6,4))
sns.countplot(x=df['Gender'])
plt.title("Phân bố giới tính")
plt.xlabel("Giới tính (0: Nữ, 1: Nam)")
plt.ylabel("Số lượng")
plt.show()

# Vẽ biểu đồ phân bố độ tuổi
plt.figure(figsize=(6,4))
sns.histplot(df['Age [years]'], bins=10, kde=True)
plt.title("Phân bố độ tuổi")
plt.xlabel("Tuổi")
plt.ylabel("Số lượng")
plt.show()

# Vẽ biểu đồ liên quan giữa chiều cao và cân nặng
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['Height [cm]'], y=df['Weight [kg]'], hue=df['Gender'])
plt.title("Chiều cao vs. Cân nặng")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Cân nặng (kg)")
plt.show()

# Vẽ biểu đồ tương quan giữa huyết áp và SpO2
plt.figure(figsize=(6,4))
sns.scatterplot(x=df['Blood pressure [mmHg]'], y=df['SpO2 [%]'], hue=df['Motion'])
plt.title("Mối quan hệ giữa huyết áp và SpO2")
plt.xlabel("Huyết áp")
plt.ylabel("SpO2")
plt.show()
```
